Pendekatan analitis dalam proyek ini didesain untuk membangun dan mengevaluasi model prediksi harga mobil bekas secara sistematis, dengan langkah-langkah sebagai berikut:

1. Pemahaman dan Persiapan Data

### E. Metrik Evaluasi

Untuk mengukur performa model dalam memprediksi harga mobil bekas, digunakan tiga metrik utama yang umum dipakai dalam permasalahan regresi:

1. Root Mean Squared Error (RMSE)  
Mengukur rata-rata selisih kuadrat antara nilai aktual dan prediksi, kemudian diakarkan. RMSE sensitif terhadap error besar, sehingga cocok untuk mengidentifikasi outlier atau prediksi ekstrem yang jauh meleset. Semakin rendah nilai RMSE, semakin akurat model.

2. Mean Absolute Error (MAE)  
Mengukur rata-rata selisih absolut antara nilai aktual dan nilai prediksi tanpa memperhitungkan arah kesalahan. MAE lebih mudah diinterpretasikan karena menggunakan satuan yang sama dengan target. MAE memberikan gambaran kesalahan rata-rata umum dalam prediksi.

3. Mean Absolute Percentage Error (MAPE)  
Mengukur rata-rata persentase error terhadap nilai aktual. MAPE sangat berguna untuk memahami seberapa besar kesalahan model dalam konteks relatif. Dalam proyek ini, MAPE digunakan sebagai acuan utama untuk menilai seberapa jauh prediksi meleset secara proporsional.

```python
# Pemeriksaan Kolom "Type"
# Apakah ada nilai yang sama antara kolom Type dan Make
type_equals_make = df[df["Type"].str.lower() == df["Make"].str.lower()]
print("- Nilai Type sama dengan Make:", type_equals_make.shape[0])
print("- Contoh:", type_equals_make["Type"].unique())
```

```python
# Pemeriksaan Kolom "Region"
# Apakah seluruh nama area sudah menggunakan huruf kapital dengan benar
capitalized_check = df["Region"].apply(lambda x: str(x).istitle())
print("- Seluruh nama area sudah memiliki huruf kapital yang benar:", capitalized_check.all())

# Apakah ada nama area yang memiliki angka (apabila hasilnya "True" maka ditemukan data anomali) 
contains_digits = df['Region'].apply(lambda x: any(char.isdigit() for char in str(x)))
print("- Ada nama area dengan angka:", contains_digits.any())

# Apakah ada spasi pada nama area (apabila hasilnya "True" maka ditemukan data anomali) 
whitespace_issues = df['Region'].apply(lambda x: str(x).strip() != str(x))
print("- Ada nama area yang memiliki spasi:", whitespace_issues.any())
```

#### - Pemeriksaan Data Pencilan (Outlier)

```python
# Identifikasi kolom numerikal
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Simpan hasil pada list
outlier_summary = []

# Identifikasi pencilan menggunakan IQR
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outlier_summary.append({
        "Column": col,
        "Outlier Count": outliers.shape[0],
        "Unique Outlier Values": sorted(outliers[col].dropna().unique().tolist())[:5]  # Batasi contoh
    })

# Simpan ke DataFrame
outlier_df = pd.DataFrame(outlier_summary)

# Tampilkan hasil
outlier_df
```

#### - Pembersihan Data

Untuk proses pembersihan data dan rekayasa fitur, digunakan dataset hasil duplikasi dari dataset awal yang sebelumnya digunakan.

```python
# Duplikasi Dataset
df_model = df.copy()
```

1. Identifikasi Data Anomali

```python
# Cek awal jumlah baris
print(f"Jumlah awal data: {df_model.shape[0]}")

# 1. Hapus baris di mana 'Type' sama dengan 'Make'
df_model = df_model[df_model['Type'] != df_model['Make']]

# 2. Hapus baris dengan placeholder pada kolom 'Make' (Other, Classic)
df_model = df_model[~df_model['Make'].isin(['Other', 'Classic'])]

# 3. Hapus tahun sebelum 1970
df_model = df_model[df_model['Year'] >= 1970]

# 4. Hapus mileage ekstrem (outlier), berdasarkan batas Z-score > 3 
from scipy.stats import zscore
df_model['Mileage_zscore'] = zscore(df_model['Mileage'])
df_model = df_model[df_model['Mileage_zscore'].abs() <= 3]
df_model.drop(columns='Mileage_zscore', inplace=True)

# 5. Hapus data dengan harga tidak relevan (<5.000 SAR atau >1.000.000 SAR)
df_model = df_model[(df_model['Price'] >= 5000) & (df_model['Price'] <= 1_000_000)]

# Tampilkan jumlah akhir data
print(f"Jumlah akhir data setelah bersih: {df_model.shape[0]}")
```

### A. Pemeriksaan Korelasi Fitur Numerikal dengan Target

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

Definisi X (fitur) dan y (target) untuk fitur numerikal

```python
# Pilih Fitur dan Target
# Untuk fitur akan menggunakan kolom numerikal dan target adalah kolom "Price"
features = ['Year', 'Mileage', 'Engine_Size']
target = 'Price'

X = df_model[features]
y = df_model[target]
```

```python
# Tambahkan Konstanta (Intercept) untuk OLS
X_const = sm.add_constant(X)  
```

```python
# Jalankan OLS 
model = sm.OLS(y, X_const).fit()
print(model.summary())
```

Interpretasi OLS:
1. F-Test (F-statistic)
- F-statistic = 534.6 dan Prob (F-statistic) = 9.88e-290.

- F-statistik ini menguji apakah model regresi secara keseluruhan signifikan atau tidak. Nilai p-value yang sangat kecil (jauh di bawah 0.05) menunjukkan bahwa model regresi ini signifikan secara statistik pada tingkat signifikansi 5%.

- Hipotesis:

--> H0: Tidak ada variabel independen yang berpengaruh terhadap Price (semua koefisien = 0).

--> H1: Setidaknya ada satu variabel independen yang berpengaruh terhadap Price.

- Karena p-value jauh lebih kecil dari 0.05, maka H0 ditolak. Artinya, model secara keseluruhan signifikan, dan ada hubungan yang bermakna antara setidaknya satu variabel independen (Year, Mileage, Engine_Size) terhadap variabel dependen (Price).

2. T-Test (t-statistics) untuk Setiap Koefisien
T-test mengevaluasi apakah masing-masing koefisien regresi secara individual berpengaruh signifikan terhadap Price.

- Koefisien const (intercept):
t = -22.909, p-value = 0.000
--> Karena p-value < 0.05, maka intercept signifikan secara statistik.

- Koefisien Year:
t = 23.015, p-value = 0.000
--> P-value sangat kecil, sehingga variabel Year berpengaruh signifikan terhadap harga mobil. Artinya, mobil yang lebih baru cenderung memiliki harga lebih tinggi.

- Koefisien Mileage:
t = -12.079, p-value = 0.000
--> P-value < 0.05, sehingga Mileage berpengaruh negatif secara signifikan terhadap harga. Artinya, semakin tinggi jarak tempuh, harga mobil cenderung menurun.

- Koefisien Engine_Size:
t = 30.208, p-value = 0.000
--> Ini menunjukkan bahwa Engine_Size secara signifikan mempengaruhi harga, dan arah pengaruhnya positif — mesin yang lebih besar cenderung menaikkan harga mobil.

- Hipotesis untuk t-test:

--> H0: Koefisien = 0 (tidak signifikan)

--> H1: Koefisien ≠ 0 (signifikan)

- Karena semua p-value < 0.05, maka semua variabel independen signifikan secara individual terhadap Price.

3. R-Squared
- R-squared = 0.297 → sekitar 29.7% variasi dalam harga mobil dapat dijelaskan oleh model ini.

- Meskipun model signifikan, nilai R² ini masih tergolong rendah, artinya sekitar 70% variasi harga belum dijelaskan, dan mungkin masih ada fitur penting lain (seperti merk, transmisi, kondisi, fitur interior, dll.) yang perlu dipertimbangkan agar model lebih baik.

4. Catatan Tambahan
- Skew dan kurtosis residual sangat tinggi (Omnibus = 2740, JB = 65422), menunjukkan residual tidak normal.

- Condition Number = 7.99e+07 → sangat tinggi, yang mengindikasikan adanya kemungkinan multikolinearitas atau masalah numerik (perlu dicek lebih lanjut dengan VIF).

- Durbin-Watson = 2.024 → mendekati 2, yang artinya tidak ada autokorelasi signifikan dalam residual.

Kesimpulan:
1. F-test menunjukkan bahwa model regresi ini signifikan secara keseluruhan.

2. T-test menunjukkan bahwa semua variabel (Year, Mileage, Engine_Size) signifikan secara individual dalam mempengaruhi harga mobil.

3. R-squared menunjukkan bahwa model ini hanya mampu menjelaskan sekitar 29.7% variasi harga, sehingga masih ada ruang untuk peningkatan dengan menambah fitur.

4. Beberapa asumsi klasik OLS (normalitas residual dan multikolinearitas) mungkin dilanggar, sehingga perlu dilakukan perbaikan model atau gunakan algoritma yang lebih robust.

```python
# Import semua library
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_val_score, KFold
from sklearn.compose import TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
import numpy as np
import pandas as pd

# Inisialisasi model
lr = LinearRegression()
lasso = Lasso(alpha=1.0, random_state=1)
ridge = Ridge(alpha=1.0, random_state=1)
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)
gbr = GradientBoostingRegressor(random_state=1)
en = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=1)
ada = AdaBoostRegressor(random_state=1)
cat = CatBoostRegressor(verbose=0, random_state=1)

# TransformedTargetRegressor untuk skala log
log_lr = TransformedTargetRegressor(regressor=lr, func=np.log, inverse_func=np.exp)
log_lasso = TransformedTargetRegressor(regressor=lasso, func=np.log, inverse_func=np.exp)
log_ridge = TransformedTargetRegressor(regressor=ridge, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(regressor=knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(regressor=dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(regressor=rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(regressor=xgb, func=np.log, inverse_func=np.exp)
log_gbr = TransformedTargetRegressor(regressor=gbr, func=np.log, inverse_func=np.exp)
log_en = TransformedTargetRegressor(regressor=en, func=np.log, inverse_func=np.exp)
log_ada = TransformedTargetRegressor(regressor=ada, func=np.log, inverse_func=np.exp)
log_cat = TransformedTargetRegressor(regressor=cat, func=np.log, inverse_func=np.exp)

# Daftar model
models = [log_lr, log_lasso, log_ridge, log_knn, log_dt, log_rf, log_xgb, log_gbr, log_en, log_ada, log_cat]
model_names = ['LinearRegression', 'Lasso', 'Ridge', 'KNN', 'DecisionTree', 'RandomForest',
               'XGBoost', 'GradientBoosting', 'ElasticNet', 'AdaBoost', 'CatBoost']

# Evaluasi model
scaler = RobustScaler()
crossval = KFold(n_splits=5, shuffle=True, random_state=1)

results = {
    'model': [],
    'mean_rmse': [], 'std_rmse': [],
    'mean_mae': [], 'std_mae': [],
    'mean_mape': [], 'std_mape': []
}

# Loop evaluasi
for name, model in zip(model_names, models):
    estimator = Pipeline([
        ('scaler', scaler),
        ('model', model)
    ])
    
    # RMSE
    rmse_scores = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                  scoring='neg_root_mean_squared_error')
    
    # MAE
    mae_scores = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                 scoring='neg_mean_absolute_error')
    
    # MAPE
    mape_scores = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                  scoring='neg_mean_absolute_percentage_error')
    
    # Simpan hasil
    results['model'].append(name)
    results['mean_rmse'].append(-rmse_scores.mean())
    results['std_rmse'].append(rmse_scores.std())
    results['mean_mae'].append(-mae_scores.mean())
    results['std_mae'].append(mae_scores.std())
    results['mean_mape'].append(-mape_scores.mean())
    results['std_mape'].append(mape_scores.std())
```

### E. Evaluasi Metrik - Data Pelatihan (Train Set)

```python
# Tampilkan hasil sebagai DataFrame
results_df = pd.DataFrame(results)
results_df.sort_values(by='mean_rmse', inplace=True)
results_df.reset_index(drop=True, inplace=True)
results_df
```

Nilai MAPE yang sangat tinggi pada Lasso dan ElasticNet menunjukkan risiko bias akibat regularisasi yang terlalu kuat.

Pipeline Model (Scaling dan Transformasi Target)

Selanjutnya, dilakukan pipeline model dengan scaling dan transformasi target.   
Langkahnya adalah sebagai berikut:  
1. Standarisasi fitur menggunakan RobustScaler

2. Transformasi target (price) ke skala log menggunakan TransformedTargetRegressor, lalu mengembalikannya ke skala asli (exp) saat prediksi

3. Model regresi di dalam pipeline, seperti CatBoostRegressor atau XGBRegressor

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import RobustScaler
from catboost import CatBoostRegressor
from xgboost import XGBRegressor

# Scaling
scaler = RobustScaler()

# Model: CatBoost (model terbaik)
cat_model = Pipeline([
    ('scaler', scaler),
    ('model', TransformedTargetRegressor(
        regressor=CatBoostRegressor(verbose=0, random_state=1),
        func=np.log,
        inverse_func=np.exp
    ))
])

# Model: XGBoost (runner-up)
xgb_model = Pipeline([
    ('scaler', scaler),
    ('model', TransformedTargetRegressor(
        regressor=XGBRegressor(random_state=1),
        func=np.log,
        inverse_func=np.exp
    ))
])

# Fit ke data training
cat_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)
```

### H. Hyperparameter Tuning

Setelah CatBoost terbukti memberikan hasil terbaik pada data pelatihan dan pengujian, langkah selanjutnya adalah melakukan hyperparameter tuning untuk mengoptimalkan kinerja model lebih lanjut. Tuning ini bertujuan untuk menemukan kombinasi parameter terbaik agar model menghasilkan prediksi yang lebih akurat dan stabil, sekaligus meminimalkan risiko overfitting.

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.compose import TransformedTargetRegressor
import numpy as np

# Model CatBoost tanpa scaler
base_cat_model = TransformedTargetRegressor(
    regressor=CatBoostRegressor(
        loss_function='RMSE',
        random_state=1,
        verbose=0,
        early_stopping_rounds=50
    ),
    func=np.log,           # untuk target skewed
    inverse_func=np.exp
)

# Parameter grid lebih luas dan realistis
cat_param_grid = {
    'regressor__depth': [6, 8, 10],
    'regressor__learning_rate': [0.03, 0.05],
    'regressor__iterations': [500, 1000],
    'regressor__l2_leaf_reg': [1, 3, 5],
    'regressor__bagging_temperature': [0, 1, 3]
}

# Randomized Search
cat_search = RandomizedSearchCV(
    estimator=base_cat_model,
    param_distributions=cat_param_grid,
    n_iter=15,                        # lebih banyak iterasi pencarian
    scoring='neg_root_mean_squared_error',
    cv=5,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Fit ke data training
cat_search.fit(X_train, y_train)

# Model terbaik
best_cat_model = cat_search.best_estimator_
```

### I. Evaluasi Metrik - Setelah Hyperparameter Tuning

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

def evaluate(y_true, y_pred, model_name):
    return {
        "Model": model_name,
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),  
        "MAE": mean_absolute_error(y_true, y_pred),
        "MAPE (%)": mean_absolute_percentage_error(y_true, y_pred) * 100
    }

# Evaluasi hasil tuning
y_pred_cat_tuned = best_cat_model.predict(X_test)
results = [evaluate(y_test, y_pred_cat_tuned, "CatBoost Tuned")]
pd.DataFrame(results)
```

### M. Validasi Silang (Cross Validation)

Setelah model dilatih dan diuji pada data training dan test set, langkah selanjutnya adalah melakukan cross-validation untuk mengevaluasi kestabilan dan kemampuan generalisasi model secara menyeluruh. Validasi silang membagi data ke dalam beberapa subset (fold) dan melatih serta menguji model secara bergantian, sehingga hasil evaluasi tidak hanya bergantung pada 1 pembagian data saja. Tahap ini penting untuk memastikan bahwa performa model tidak hanya baik pada data tertentu, tetapi juga konsisten di berbagai skenario data yang berbeda.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.compose import TransformedTargetRegressor
from catboost import CatBoostRegressor
import numpy as np

# Inisialisasi KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Placeholder untuk skor
rmse_scores = []
mae_scores = []
mape_scores = []

# Loop manual CV
for train_index, val_index in kf.split(X_train):
    X_tr, X_val = X_train.iloc[train_index], X_train.iloc[val_index]
    y_tr, y_val = y_train.iloc[train_index], y_train.iloc[val_index]

    # Model pipeline dengan CatBoost
    cat_model_cv = TransformedTargetRegressor(
        regressor=CatBoostRegressor(verbose=0, random_state=1),
        func=np.log,
        inverse_func=np.exp
    )

    # Fit dan prediksi
    cat_model_cv.fit(X_tr, y_tr)
    y_pred = cat_model_cv.predict(X_val)

    # Hitung metrik
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    mae = mean_absolute_error(y_val, y_pred)
    mape = mean_absolute_percentage_error(y_val, y_pred)

    # Simpan ke list
    rmse_scores.append(rmse)
    mae_scores.append(mae)
    mape_scores.append(mape)

# Cetak hasil
print("CatBoost Cross-Validation")
print(f"RMSE: {np.mean(rmse_scores):,.2f} ± {np.std(rmse_scores):,.2f}")
print(f"MAE : {np.mean(mae_scores):,.2f} ± {np.std(mae_scores):,.2f}")
print(f"MAPE: {np.mean(mape_scores)*100:.2f}% ± {np.std(mape_scores)*100:.2f}%")
```

Hasil cross-validation menunjukkan bahwa model CatBoost memiliki performa yang cukup stabil, dengan RMSE rata-rata sebesar 32.428 (±5.042) dan MAE sebesar 14.393 (±444). Nilai MAPE sebesar 18.95% (±0.70%) berada dalam kisaran yang masih dapat diterima untuk prediksi harga, khususnya di industri otomotif, di mana akurasi prediksi dalam rentang MAPE 10%–20% masih dianggap wajar (Kohavi & Longbotham, 2017; Lewis, 1982). Deviasi standar yang rendah juga mengindikasikan konsistensi performa model antar fold, yang menjadi indikator stabilitas model dalam proses validasi silang.
